__(1: 20 pts)__ Twenty buildings in a city have been selected that will have microwave receivers located on their rooftops. Each receiver will be served by an assigned microwave transmitter that will be located on the rooftop of one of these buildings. Each transmitter will operate on a different frequency to limit signal interference, but the signal from a transmitter to a receiver on the rooftop of a building can only reach 60 units of distance because of signal attenuation, and a transmitter should not be closer than 20 units of distance from a receiver operating on the same frequency as its assigned. The coordinates of the buildings can be generated by running the following code:

```
using Random, Statistics
Random.seed!(3894)
P = 200*rand(20,2)
```

Determine (a) the buildings where transmitters should be located and (b) to which transmitter each receiver should be assigned.

_Your Summary:_ 

I have taken reference of set covering concepts from MIP3 notebook.

````julia
using Random, Statistics
Random.seed!(3894)
P = 200*rand(20,2)  # Co-ordinates of Receivers
````

````julia
function dist(x,P)
    distance = sqrt.(sum((x' .- P).^2, dims=2))
end
D = vcat([dist(P[i,:],P) for i in 1:size(P,1)]'...)

# the distance bwtween Transmitter and Receiver must be in between 20 and 60
A = 20 .< D .< 60  
````

````julia
using JuMP, Cbc
function setcover(A)
    M = 1:size(A,1)
    N = 1:size(A,2)
    model = Model(Cbc.Optimizer)
    @variable(model, x[1:length(N)], Bin)
    @objective(model, Min, sum(x[i] for i in N))
    @constraint(model, [j ∈ M], sum(A[j,i]*x[i] for i in N) >=1)
    set_optimizer_attribute(model, "logLevel", 0)
    set_optimizer_attribute(model, "seconds", 60.0)
    optimize!(model)
    println(solution_summary(model).termination_status)
    return findall(value.(x) .> 0)
end

idx = setcover(A)
println(string(length(idx)), " index to locate transmitters\n", idx)
````

````julia
using Plots
scatter(P[:,1], P[:,2], label="City", markersize=2, aspect_ratio = :equal, title="Transmitter and City Plot")
scatter!(P[idx,1], P[idx,2], label="Transmitter", legend=:outerright)
````

````julia
println("Co-ordinates of 8 Transmitters-->")
[(P[id,1], P[id,2]) for id in idx]
````

b) To which transmitter each receiver should be assigned?

````julia
# A matrix is symmetric 

for i in idx
    j = unique(findall(A[i,:].== 1))  # checking all location which satisfies the constraint condition
end
````

__(2: 15 pts)__ A supplier provides a critical biological material to a facility that produces a pharmaceutical product. The material is transported 1278 miles to the facility via truckload, and the cost for each truckload is \\$3.55 per mile, which is independent of how much is transported. Each truckload can carry up to 25 tons and 2750 cubic feet of material. Each unit of raw material costs \$15000, occupies 300 cubic feet and weighs 350 pounds. Forecasted demand (in units per week) is 7, 5, 14, 6, 1, 2, 12, 3, 27, 8, 9, 8, and 12. A minimum of two units and a maximum of 26 units can be stored at the plant. Currently, eight units are in storage, and the same number should be in storage at the end of the planning horizon. Assuming that the annual inventory carrying rate is three, determine the best transportation plan over the planning horizon.

_Your Summary:_ 

I took reference from the Dr.Kay's example solution (Ex-2 TruckLoad Transport) from notebook MIP4: Production Inventory Planning.

````julia
rTL = 3.55           # cost per mile
d = [7,5,14,6,1,2,12,3,27,8,9,8,12]
distance = 1278      # distance in miles given 
cTL = rTL*distance   # total cost per TL
nCube = floor(2750/300)
nWt = floor(25*2000/350)

# Min of the possible unit based on the cube and weight constraints
nTL = min(nCube, nWt)
````

````julia
u =  [nTL]    #capacity at each stage units
cp = [15000] # variable production cost $/unit
ck = [cTL]   # fixed production cost at each stage ($)
h =  3/(365.25/7)  # inventory carrying rate per week ($/week)
ci = cumsum(cp + ck./u)*h   # inventory carring cost at each stage($/unit)

y0 = 8   # min inventory at each stage
yt = y0  # max inventory at each stage
m = length(u) # number of production stage
t = length(d) # number of period of production

Ylb = [y0 repeat([2],m,t-1) yt]          # same LB and UB to fix inital and final inventory levels
Yub = [y0 repeat([26],m,t-1) yt]         # max inventory storage level
````

````julia
using JuMP, Cbc
model = Model(Cbc.Optimizer)
@variable(model, 0<=X[i=1:m, j=1:t])
@variable(model, Ylb[i,j] <= Y[i=1:m, j=1:t+1] <= Yub[i,j])
@variable(model, Z[i=1:m, j=1:t], Int)
@objective(model, Min, 
    sum(ci[i]*Y[i,j] for i=1:m, j=2:t+1 ) + sum(ck[i]*Z[i,j] for i=1:m, j=1:t ))
for j = 1:t
    for i = 1:m-1                                      # (a) flow-balance all but last stage
        @constraint(model, -X[i,j] + X[i+1,j] - Y[i,j] + Y[i,j+1] == 0 )   
    end                                                # (b) flow-balance for last stage
    @constraint(model, -X[m,j] - Y[m,j] + Y[m,j+1] == -d[j] )              
    for i = 1:m                                        # (c) production capacity
        @constraint(model, X[i,j] <= u[i]*Z[i,j] )                         
    end
end
set_optimizer_attribute(model, "logLevel", 0)
set_optimizer_attribute(model, "seconds", 30.0)
optimize!(model)
println(solution_summary(model).termination_status)
TCᵒ, Xᵒ, Yᵒ, Zᵒ = objective_value(model), value.(X), value.(Y), round.(value.(Z))
````

````julia
# Report results
println(stdout,"\nTotal transport and inventory cost: ", TCᵒ)
println(stdout,"\nUnits transported/Demand: "), flush(stdout)
display(Matrix([Xᵒ; d']))
println(stdout,"\nInventory level: "), flush(stdout)
display(Matrix(Yᵒ))
println(stdout,"\nTL per week: "), flush(stdout)
display(Int.(Matrix(Zᵒ)))
````

__(3: 15 pts)__ Continuing with Question 2 from _HW6_ (using the same flow and routing data), determine a machine layout that minimizes the total cost for the following different set of site locations:



The sites are now located along a circular conveyor that can only travel in a clockwise direction.

_Your Summary:_ 

Using Graphs module a distance to distance matrix i.e each node i to ending node j formulated. As the machine layout is simple not convoluted as a network model a linear distance calulated from each node i to ending node j. Route of 6 products is similar to HW-6 Q-2. Based on intial allocation vector α, a total cost calculated which further minimised using subsequent sdpi and multisdpi function. A multistart procedure is used to utilised the best allocation as the initial allocation of the problem.


````julia
using LightGraphs, SimpleWeightedGraphs
i = [1,8,7,6,5,4,3,2]    #index of starting node of arc
j = [8,7,6,5,4,3,2,1]    # index of ending node of arc
w = [26, 7, 4, 5, 16, 10, 11, 8]   #weight(distance) associated with each erc

G = SimpleWeightedGraph(i, j, w) # create a undirected network

using GraphPlot, Random
Random.seed!(52)
gplot(G, nodelabel=1:nv(G), edgelabel=[weights(G)[src(e),dst(e)] for e in edges(G)])
````

````julia
D = Matrix(adjacency_matrix(G)) # Distance to Distance matrix i.e each node i to ending node j distance
````

````julia
frequency = [242, 472, 351, 82, 118, 735] # given in Q-2 HW6

# Route of 6 product in Machine layout as per the Q-2 HW6
route = [[3,6,5,8,6,4], 
    [5,3,4,1,7,5], 
    [8,4,2,6,5,3,1,7],
    [6,1,4,7,3,5,8,2], 
    [5,8,2,6,5,3,1,7], 
    [7,6,4,6,4]]
n = maximum([maximum(i) for i in route])
println("Maximum number of Machines: ",n)
````

````julia
# Weight Matrix of flow volume using route and frequency
W = zeros(n,n)
for i in 1:length(frequency), (j,k) in zip(route[i][1:end-1],route[i][2:end])
    @show j,k
    W[j,k] += frequency[i]
end
W   # Machine to machine matrix
````

````julia
# Let α be the assignment vector
α = [1,2,3,4,5,6,7,8]
TC = sum(W[α,α].*D)
print("\n [INFO] Total cost associated with the initial machine layout = ",TC )
````

#### Improving / Minimizing the total cost associated with the layout using SDPI method

````julia
function sdpi(α°, W,D)
    TC°, n, cnt = sum(W[α°,α°]*D), length(α°), 1
    done = false
    while !done
        done = true
        α° = copy(α°)
        for i = 1:n-1, j=i+1:n
            cnt += 1
            α = copy(α°)           # Work on copy of current assignment
            α[[j,i]] .= α[[i,j]]   # Pairwise interchange of i and j
            TC = sum(W[α,α].*D)    # Calc. TC of interchange
            if TC < TC°            # Improvement found
                done = false
                TC°, α° = copy(TC), copy(α)
            end
        end
    end
    return (TC = TC°, α = α°, cnt = cnt)   # Named tuple used to pass multiple outputs
end

# A multistart procedure can be utilized to find best of n runs of sdpi
function multisdpi(W,D)
    Total_Cost = []
    TCᵒ, αᵒ, n, cnt = Inf, [], size(W,1), 0
    for i = 1:n
        res = sdpi(randperm(n),W,D)
        cnt += res.cnt
        if res.TC < TCᵒ
            (TCᵒ,αᵒ) = res
        end
        println("\n(",i,") ",res.TC,": ",res.α,"\n")
        append!(Total_Cost, res.TC)
    end
    return (TC = TCᵒ, α = αᵒ, cnt = cnt, Total_Cost)
end
Random.seed!(12546)
_,alpha,_,TC = multisdpi(W,D)
````

````julia
using Plots
x = 1:length(α)
plot(x, TC, xlim = (0,9), ylim=(1000,6500), lw=2, label=false, title ="Total cost Variation")
xlabel!("Length of α°")
ylabel!("Total Cost")
scatter!([7], [2034], label="Min Cost")
````

````julia
println("The machine layout for minimum cost: ", alpha)
````

__(4: 15 pts)__ Determine the minimum number of containment devices, each with a size capacity of twenty-five that will be needed to store three-thousand samples. The size of each sample is the equally likely integer from ten to twenty equally generated using a random number seed of 254. 

_Your Summary:_ 

I have used Heuristics approch to solve the Bin problem as Cbc is unable to solve the problem. With reference to 6-Comb_Opt-2 notebook, it is suggested that some time it is difficult to solve bin problem using MIP and in such scenario use of a heuristic is helpful.

````julia
using Random
m = 3000       #number of sampls (trial)
M = 1:m
V = 25        #size capacity of each containment devices
Random.seed!(254)
v = rand(10:20,m)   # size of each sample
LB = ceil(sum(v)/V)
print("Lower Bound: ",LB)
````

````julia
function firstfit(v,V)
    # Put first object into first bin
    B = [[1]]                               # Array of arrays
    b = [v[1]]                              # Array of scalars
    # Add remaining objects to bins
    for i = 2:length(v)
        idx = findfirst(b .+ v[i] .<= V)    # Find first bin object fits into
        if ~isnothing(idx)                  # Found existing bin
            b[idx] += v[i]
            push!(B[idx],i)
        else                                # Create new bin for object
            push!(b,v[i])
            push!(B,[i])
        end
    end
    return B
end
B = firstfit(v,V)
````

````julia
function multifirstfit(v,V)
    m = length(v)
    TC, B = Inf, 0
    for i = 1:m
        Bi = firstfit(v[randperm(m)],V)
        if length(Bi) < TC
            TC, B = length(Bi), Bi
        end
    end
    return B
end
Random.seed!(43)
LB, length(multifirstfit(v,V))
````

````julia
println("Based on the Heurastic approach the, minimum number of containment devices=  ", length(multifirstfit(v,V)))
````

````julia
using DataFrames, CSV, MLJ
trainFile = "ML-3-Titanic-train.csv"
trainDF = DataFrame(CSV.File(trainFile))
first(trainDF,5)
````

#### EDA on train dataset

````julia

# Rename the long character into a smaller version 
rename!(trainDF, "Siblings/Spouses Aboard"=>:Sibsp, "Parents/Children Aboard"=>:Parch)

# As name is not sigificantly contribution to the training of data, we are droping from the set
trainDF = select(trainDF, Not(:Name))

# Converting survival into OrderedFactor
trainDF.Survived = coerce(trainDF.Survived, OrderedFactor)

# Cosnverting sex feature from categorical
trainDF.Sex = coerce(trainDF.Sex, OrderedFactor)
trainDF.Sex = coerce(trainDF.Sex, Count)

y = trainDF.Survived
X = select(trainDF, Not(:Survived))
first(X, 5)
````

````julia
# Fitting complete dataset with an unrestricted deoth of DecisionTreeClassifier
@load DecisionTreeClassifier pkg=DecisionTree verbosity=0
model = MLJDecisionTreeInterface.DecisionTreeClassifier()
mach = machine(model, X, y)
fit!(mach, verbosity=0)
mach.fitresult
````

````julia
# Find the best depth of DecisionTreeClassifier out of total 24 depth
using Random
Random.seed!(4453)
maxDepth = 24
n = 25
depth°, acc°, mach° = 0, 0, nothing
for i=1:maxDepth
    model = MLJDecisionTreeInterface.DecisionTreeClassifier(max_depth = i)
    mach = machine(model, X, y)
    for j=1:n
        i,j
        train, val = partition(eachindex(y), 0.7, shuffle=true)   # 70% train data and 30% test data
        fit!(mach, rows=train, verbosity=0)
        ŷ = predict_mode(mach, rows=val)
        acc = accuracy(ŷ, y[val])
        if acc>acc°
            acc°,depth°, mach° = acc, i, mach
        end
    end
end
depth°, acc°
````

###### DecisionTreeClassifier with restricted depth, giving depth=3 as a best possible depth for maximum accuracy
