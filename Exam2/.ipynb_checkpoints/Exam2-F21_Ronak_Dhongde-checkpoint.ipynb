__(1: 20 pts)__ Twenty buildings in a city have been selected that will have microwave receivers located on their rooftops. Each receiver will be served by an assigned microwave transmitter that will be located on the rooftop of one of these buildings. Each transmitter will operate on a different frequency to limit signal interference, but the signal from a transmitter to a receiver on the rooftop of a building can only reach 60 units of distance because of signal attenuation, and a transmitter should not be closer than 20 units of distance from a receiver operating on the same frequency as its assigned. The coordinates of the buildings can be generated by running the following code:

```
using Random, Statistics
Random.seed!(3894)
P = 200*rand(20,2)
```

Determine (a) the buildings where transmitters should be located and (b) to which transmitter each receiver should be assigned.

_Your Summary:_ 

__1.__ We first Generate the coordinates for the 20 buildings. Then using distance formula we build the D matrix with vcat.
We have a condition that the distance between transmitter and receiver must be no greater than 60 units and no smaller than 20. Using this condition we get the 20x20 binary matrix indicating the building distances which comply with the constraint.

````julia
using Random, Statistics
Random.seed!(3894)
P = 200*rand(20,2)
dist1(x,P) = sqrt.(sum((x' .- P).^2, dims =2 ))       #using distance formula to find the distances between each 20 biuldings
D = vcat([dist1(P[i,:],P) for i in 1:size(P,1)]'...)  #using vcat to return a matrix "D"
A = 20 .<= D .<= 60                                   #using the given condition to filter out the buildings which comply
````

__2.__ Then we use a Cbc solver to define and solve the problem. We set objective of minimizing the number of transmitters.
We define the constraint that each transmitter should be assigned atleast one receiver.
Proceeding with optimizing the model, we then print the optimal solution, which gives us the building numbers where the transmitters should be installed

````julia
using JuMP, Cbc
function setcover(A)
    M = 1:size(A,1)                                               #defining scope for transmitters
    N = 1:size(A,2)                                               #defining scope for receivers
    model = Model(Cbc.Optimizer)                                  #initiating the use of model
    @variable(model, x[1:length(N)], Bin)                         #using the binary variable as a deciding factor
    @objective(model, Min, sum(x[i] for i in N))                  #constructing the objective function
    @constraint(model, [j in M], sum(A[j,i]*x[i] for i in N) >= 1)#constraint for assigning atleast one receiver 
    set_optimizer_attribute(model, "loglevel", 0)                 
    set_optimizer_attribute(model, "seconds", 60.00)              #max compilation time
    optimize!(model)        
    println(solution_summary(model).termination_status)           #printing the solution
    return findall(value.(x) .> 0)
end
Z = setcover(A)
````

__3.__ Plotting the buildings and transmitters

````julia
using Plots
scatter(P[:,1], P[:,2], label = "Building",title = "\nMap of buildings and transmitters", markersize = 2, aspect_ratio =:equal)   #plotting the buildings
scatter!(P[Z,1], P[Z,2], label= "Building with transmitter", legend =:outerright)   #plotting the buildings with transmitters
````

__4.__ We now print out the coordinates of the transmitters using a for loop by iterating through our solution Z

````julia
for i in Z
    print(P[i,1], "   " , P[i,2])
    print("\n")
end
````

__5.__ After finding the coordinates, we need to zip the coordinates of buildings with transmitters with their counterpart receiver. We again iterate through a for loop taking into account the A matrix where we had implied the condition of distance.

````julia
for i in Z
    idx = findall(A[:,i] .> 0)
    print("\n\nFor transmitter mounted on building No.: ",i)
    print("\nThe receivers are located at: ",idx)
end
````

__(2: 15 pts)__ A supplier provides a critical biological material to a facility that produces a pharmaceutical product. The material is transported 1278 miles to the facility via truckload, and the cost for each truckload is \\$3.55 per mile, which is independent of how much is transported. Each truckload can carry up to 25 tons and 2750 cubic feet of material. Each unit of raw material costs \$15000, occupies 300 cubic feet and weighs 350 pounds. Forecasted demand (in units per week) is 7, 5, 14, 6, 1, 2, 12, 3, 27, 8, 9, 8, and 12. A minimum of two units and a maximum of 26 units can be stored at the plant. Currently, eight units are in storage, and the same number should be in storage at the end of the planning horizon. Assuming that the annual inventory carrying rate is three, determine the best transportation plan over the planning horizon.

_Your Summary:_ 

__1.__ We first formulate the demand and given data. We need to choose between the maximum weight and maximum volume each truck load can carry. The minimum of the either will then satisfy the remaining one. Thus, the maximum  units we can carry in a single TL without violating the weight and volume constraint is 9.0 units

````julia
demand = [7, 5, 14, 6, 1, 2, 12, 3, 27, 8, 9, 8, 12]  #total demand
distance = 1278                                       #distance between destination and starting point
cost_mile = 3.55                                      #transportation cost per mile
total_tlcost = distance*cost_mile                     #total fixed transportation cost
maximum_cube = floor(2750/300)                        #condition for maximum volume that can be carried in 1TL
maximum_wt = floor(25*2000/300)                       #condition for maximum weight that can be carried in 1TL
max_unit = min(maximum_cube, maximum_wt)              #choosing minimum between both conditions to satisfy both
````

__2.__ Formulating the other details like capacity, variable costs, length of period, inventory lows and highs (LB,UB)

````julia
str = [max_unit]                     #maximum capacity
cp = [15000]                         #cost of RM
ck = [total_tlcost]                  #cost for 1TL
h = 3/(365.25/7)                     #holding cost in days
ci = cumsum(cp + ck./str)*h          #inventory cost for model
y0 = [8]                             #initial inventory
yt = y0                              #final inventory
m = length(str)                      
t = length(demand)
Ylb = [y0 repeat([2],m,t-1) yt]      #LB for inventory at each period
Yub = [y0 repeat([26],m,t-1) yt]     #UB for inventory at each period
print(Ylb)
print("\n",Yub)
````

__3.__ Then we use a Cbc solver to define and solve the problem. We set objective of minimizing the total cost for transportation and inventory. Proceeding with optimizing the model, we then print the optimal solution, which gives us the Total cost of transportation, the units transported during one period, the inventory level during that period after replenishment and meeting the demand, and the truck trip count for each period. 

````julia
using JuMP, Cbc
model = Model(Cbc.Optimizer)
@variable(model, 0 <= X[i=1:m, j=1:t] )
@variable(model, Ylb[i,j] <= Y[i=1:m, j=1:t+1] <= Yub[i,j] )
@variable(model, Z[i=1:m, j=1:t], Int )   # Int instead of Bin since can have multiple TL/wk
# Note: Unit prod cost not included in obj since does not impact soln, used only for inv cost
@objective(model, Min, 
    sum(ci[i]*Y[i,j] for i=1:m, j=2:t+1 ) + sum(ck[i]*Z[i,j] for i=1:m, j=1:t ) )
for j = 1:t
    for i = 1:m-1                                      # (a) flow-balance all but last stage
        @constraint(model, -X[i,j] + X[i+1,j] - Y[i,j] + Y[i,j+1] == 0 )   
    end                                                # (b) flow-balance for last stage
    @constraint(model, -X[m,j] - Y[m,j] + Y[m,j+1] == -demand[j] )              
    for i = 1:m                                        # (c) production capacity
        @constraint(model, X[i,j] <= str[i]*Z[i,j] )                         
    end
end
set_optimizer_attribute(model, "logLevel", 0)
set_optimizer_attribute(model, "seconds", 30.0)
optimize!(model)
println(solution_summary(model).termination_status)
TCáµ’, Xáµ’, Yáµ’, Záµ’ = objective_value(model), value.(X), value.(Y), round.(value.(Z))
````

````julia
# printing out the final results
println(stdout,"\nTotal transport and inventory cost: ", TCáµ’)
println(stdout,"\nUnits transported/Demand: "), flush(stdout)
display(Matrix([Xáµ’; demand']))
println(stdout,"\nInventory level: "), flush(stdout)
display(Matrix(Yáµ’))
println(stdout,"\nTL per week: "), flush(stdout)
display(Int.(Matrix(Záµ’)))
````

__1.__ We prepare the sdpi function which can be passed under with a counter, W matrix and D matrix

````julia
function sdpi(alpha,W,D)
    TCáµ’, n, cnt = sum(W[alpha,alpha].*D), length(alpha), 1
    println(TCáµ’,": ",alpha)
    done = false
    while !done
        done = true                # Assume no improvement will be found
        alpha = copy(alpha)              # Copy as current in case improvement found
        for i = 1:n-1, j = i+1:n
            cnt += 1
            alpha1 = copy(alpha)           # Work on copy of current assignment
            alpha1[[j,i]] .= alpha1[[i,j]]   # Pairwise interchange of i and j
            TC = sum(W[alpha1,alpha1].*D)    # Calc. TC of interchange
            if TC < TCáµ’            # Improvement found
                done = false
                TCáµ’, alpha = copy(TC), copy(alpha1)
            end
        end
        println(TCáµ’,": ",alpha)       # Output best interchange found at each step
    end
    return (TC = TCáµ’, alpha1 = alpha, cnt = cnt)   # Named tuple used to pass multiple outputs
end
````

__(4: 15 pts)__ Determine the minimum number of containment devices, each with a size capacity of twenty-five that will be needed to store three-thousand samples. The size of each sample is the equally likely integer from ten to twenty equally generated using a random number seed of 254. 

_Your Summary:_ 

__1__. We solve this bin packing problem using heuristics and use firstfit and multifit approaches to examine the difference in the bins

Defining all the parameters and finding the lower bound of total bins required by summing up the volumes of all elements and dividing them by total volume of the bin

__Starting with firstfit__

````julia
function firstfit(v,V)
    # Put first object into first bin
    B = [[1]]                               # Array of arrays
    b = [v[1]]                              # Array of scalars
    # Add remaining objects to bins
    for i = 2:length(v)
        idx = findfirst(b .+ v[i] .<= V)    # Find first bin object fits into
        if ~isnothing(idx)                  # Found existing bin
            b[idx] += v[i]
            push!(B[idx],i)
        else                                # Create new bin for object
            push!(b,v[i])
            push!(B,[i])
        end
    end
    return B
end
````

````julia
using Random
m = 3000
M = 1:m
V = 25
Random.seed!(254)
v = rand(10:20,m)
LB = ceil(sum(v)/V)
A = firstfit(v,V)
LB, length(A)
````

````julia
[sum(v[A[i]]) for i=1:length(A)]               #total volume of bin filled with items
````

__Multifirstfit approach:__

````julia
function multifirstfit(v,V)
    m = length(v)
    TC, B = Inf, 0
    for i = 1:m
        Bi = firstfit(v[randperm(m)],V)
        if length(Bi) < TC
            TC, B = length(Bi), Bi
        end
    end
    return B
end
Random.seed!(254)
LB, length(multifirstfit(v,V))
````

````julia
[v[B[i]] for i=1:length(B)]          #volume of items in each bin
````

````julia
[sum(v[B[i]]) for i=1:length(B)]               #total volume of bin filled with items
````
